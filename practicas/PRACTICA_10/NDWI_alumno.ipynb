```python
# Función de GEE predefinida para añadir capas con Folium

def add_ee_layer (self, ee_imagen, parametros, nombre):
    map_id = ee.Image(ee_imagen).getMapId(parametros)
    folium.raster_layers.TileLayer(
        tiles=map_id['tile_fetcher'].url_format, 
        attr = 'Adrián y Fedor. Tecnologias de la Informacion Geografica', #Indicar el nombre de los componentes del grupo
        name = nombre,
        overlay = True, # Permite la superposición
        control = True # Permite que aparezca la capa o no en el layer control
        ).add_to(self)

# Con esto creamos un método propio llamado addLayer que llamara a la función add_ee_layer
folium.Map.addLayer = add_ee_layer
```

<h3 align="center">Índice de Agua de Diferencia Normalizada (NDWI). McFeeters, 1996</h3>
<h3 align="center">Índice de Agua de Diferencia Normalizada Modificado (MNDWI). Xu, 2006</h3>
</div>


```python
import folium
from folium import plugins
import ee
import branca
import branca.colormap as cm
ee.Initialize()
```

```python
# Colección de imágenes de Sentinel con una nubosidad inferior al 20% y la medía calculada de los pixels
landsat= (ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")
                  .filterDate('2025-03-21','2025-06-21')
                  .filter(ee.Filter.lt('CLOUD_COVER',20)).mean()) #el filtro de nubosidad varía según imagen

sentinel= (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
                  .filterDate('2025-03-21','2025-06-21')
                  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE',20)).mean()) #el filtro de nubosidad varía según imagen

def NDWI(image):
    NDWI = image.expression('float((GREEN - NIR)/(GREEN + NIR))',{
        'NIR' : image.select("SR_B5").multiply(2.75e-05),     # Escala según el catalogo de datos de GEE para sentinel2
        'GREEN' : image.select("SR_B3").multiply(2.75e-05),
    })
    return NDWI

def MNDWI(image):
    MNDWI = image.expression('float((GREEN - SWIR)/(GREEN + SWIR))',{
        'SWIR' : image.select("B11").multiply(0.0001),     # Escala según el catalogo de datos de GEE para sentinel2
        'GREEN' : image.select("B3").multiply(0.0001),
    })
    return MNDWI

sld_intervalos = """
    <RasterSymbolizer>
        <ColorMap type="intervals" extended="false" >
            <ColorMapEntry color="#FF0000" quantity="-0.3" label="Superficies sin agua"/>
            <ColorMapEntry color="#FFFF00" quantity="0" label="Sequía moderada" />
            <ColorMapEntry color="#008000" quantity="0.2" label="Inundación, humedad" />
            <ColorMapEntry color="#0000FF" quantity="1" label="Superficie de agua" />
        </ColorMap>
    </RasterSymbolizer>"""

# Llamada a las funciones
NDWI_res = NDWI(landsat)
MNDWI_res = MNDWI(sentinel)

# Exportar imagen a Google Drive. Configuración parámetros de salida
geometry = ee.Geometry.Rectangle([43.207593, 12.002512, 43.076828, 12.190309]) # Coordenadas geográficas minima (longitud, latitud) y maxima de la imagen a exportar
task_config = {
        'region': geometry,
        'fileFormat': 'GeoTIFF',
        'fileNamePrefix': 'FedorAdrian_NDWI',
        'folder':'gee-datos',
        'scale': 20,      # Resolución en metros del pixel, siempre la resolución más desfavorable de las bandas utilizadas 
        'crs': 'EPSG:4326',
        'image': NDWI_res,
        'description': 'Imagen procesada NDWI'
    }
ee.batch.Export.image.toDrive(**task_config).start()

mapa_NDWI = folium.Map(location=[43.147114, 12.101056], zoom_start = 11) #Las coordenadas son en Latitud y Longitud
mapa_NDWI.addLayer(NDWI_res.sldStyle(sld_intervalos),None,'NDWI')

mapa_MNDWI = folium.Map(location=[37.193668, 9.852831], zoom_start = 12)
mapa_MNDWI.addLayer(MNDWI_res.sldStyle(sld_intervalos),None,'MNDWI')

folium.LayerControl().add_to(mapa_NDWI)
folium.LayerControl().add_to(mapa_MNDWI)


mapa_NDWI



 
```
